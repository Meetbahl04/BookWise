```python
from scipy.sparse import csr_matrix
book_sparse = csr_matrix(books_pivot)
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(n_neighbors = 6, metric = 'cosine')
model.fit(book_sparse)
distances, suggestions = model.kneighbors(books_pivot.iloc[47, :].to_numpy().reshape(1, -1))
```

```python
import numpy as np

def book_recommend(book_name):
    book_id = np.where(books_pivot.index == book_name)[0][0]
    distances, suggestions = model.kneighbors(books_pivot.iloc[book_id, :].to_numpy().reshape(1, -1))
    
    print(f"Suggestions for the book '{book_name}':")
    for i, book_index in enumerate(suggestions[0], start=1):
        print(f"{i}. {books_pivot.index[book_index]}")
```

```python
book_recommend('Animal Farm')
```

```python
def precision_recall(recommended_books, actual_books):
    relevant_recommendations = len(set(recommended_books) & set(actual_books))
    precision = relevant_recommendations / len(recommended_books)
    recall = relevant_recommendations / len(actual_books)
    return precision, recall

actual_books = ["1984", "Brave New World", "Fahrenheit 451"]  # Replace with real ground truth data
recommended_books = [books_pivot.index[i] for i in suggestions[0]]

precision, recall = precision_recall(recommended_books, actual_books)
print(f"Precision: {precision}, Recall: {recall}")
```
